```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
    
plot_correlation_map( df )
```

```python
Title_Dictionary = {'Capt': 'Officer', 'Col': 'Officer', 'Major': 'Officer', 'Dr': 'Officer', 'Rev': 'Officer', 'Jonkheer': 'Royalty', 'Don': 'Royalty', 'Sir': 'Royalty', 'Lady': 'Royalty', 'the Countess': 'Royalty', 'Dona': 'Royalty', 'Mme': 'Miss', 'Mlle': 'Miss', 'Miss': 'Miss', 'Ms': 'Mrs', 'Mr': 'Mrs', 'Mrs': 'Mrs', 'Master': 'Master'}
def get_title_from_name(Name):
    name = [w.replace('.', '').lower() for w in Name.split()]
    for k in Title_Dictionary:
        if k.lower() in name:
            return Title_Dictionary[k]
    return 'Mrs'
```

```python
df['Title'] = df['Name'].apply(get_title_from_name)
df = df.drop('Name', axis=1)
```

create a new column and Visualization the correlation between the new column and  other features and collection the titles into more useful ones.

```python
df["Survived"] = df["Survived"].replace({"Yes": 1, "No": 0})
df["Sex"] = df["Sex"].replace({"male": 1, "female": 0})
df["Embarked"] = df["Embarked"].replace({"C": 1, "Q": 0, "S": 2})
df["Title"] = df["Title"].replace({v:k for k,v in enumerate(set(Title_Dictionary.values()))})
```

```python
df = df.drop(['SibSp', 'Parch', 'Ticket'], axis=1)
df.info()
```

```python
from sklearn.model_selection import train_test_split

Y = df['Survived']
X = df.drop('Survived', axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.20)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import *
Y_pred = model.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred)
cm
```

```python
tn, fp, fn, tp = cm.ravel()
(tn, fp, fn, tp)
```

```python
ax = sns.heatmap(pd.DataFrame(cm), annot=True)
ax.set_xlabel("Predicted")
ax.set_ylabel("Real")
```
